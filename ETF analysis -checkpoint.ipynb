### Monte Carlo Simulation for Portfolio Risk Assessment
##### Problem Statement
- A financial analyst aims to assess the potential risk for a portfolio comprising 11 stocks from the **CPSE ETF (Central Public Sector Enterprises ETF)**. 
- The objective is to predict the future returns of this portfolio using **Monte Carlo Simulation**.

#### Why Monte Carlo Simulation?
Instead of relying solely on historical data, this method generates thousands of potential future outcomes for stock prices. By considering historical 
volatility and returns, it enables:
- Estimation of the probability distribution of portfolio returns.
- Calculation of **Value at Risk (VaR)** to quantify potential downside risks.

#### About CPSE ETF
The **CPSE ETF** is a sectoral **Exchange-Traded Fund (ETF)** that invests in a diversified basket of public sector enterprises (PSEs) in **India**, 
with a focus on the **energy sector**.

#### Key Characteristics

- **Government-Backed**: The ETF includes prominent enterprises with significant influence in India's energy infrastructure and market.

```python
# Step 1: Download data
stocks = [ "POWERGRID.NS",  
    "NTPC.NS",      
    "BEL.NS",  
    "ONGC.NS",
    "COALINDIA.NS",  
    "OIL.NS",      
    "NHPC.NS",  
    "COCHINSHIP.NS",
    "NBCC.NS",  
    "NLCINDIA.NS",      
    "SJVN.NS"]
data = yf.download(stocks, start='2023-12-01', end='2024-12-01')['Adj Close']
# Step 2: Calculate returns
returns = data.pct_change().dropna()
mean_returns = returns.mean()
std_devs = returns.std()

# Step 3: Monte Carlo simulation
days = 252  # Simulating for one year (252 trading days)
num_simulations = 10000
num_stocks = len(stocks)
simulated_prices = np.zeros((days, num_stocks, num_simulations))

for i, stock in enumerate(stocks):
    last_price = data[stock].iloc[-1]
    drift = mean_returns[i] - (0.5 * std_devs[i] ** 2)  # Drift for the stock
    daily_volatility = std_devs[i]
    random_shocks = np.random.normal(0, 1, (days, num_simulations))
    price_paths = last_price * np.exp(np.cumsum(drift + daily_volatility * random_shocks, axis=0))
    simulated_prices[:, i, :] = price_paths

# Step 4: Portfolio returns
weights = np.array([1/num_stocks] * num_stocks)  # Equal weights for all 11 stocks
portfolio_values = np.sum(simulated_prices * weights[:, np.newaxis], axis=1)

# Step 5: Calculate Value at Risk (VaR)
final_portfolio_values = portfolio_values[-1, :]  # Portfolio values at the end of the simulation
VaR_95 = np.percentile(final_portfolio_values, 5)  # 5th percentile (95% confidence level)
print(f"Value at Risk (5%): {VaR_95:.2f}")

# Step 6: Visualization
fig, axes = plt.subplots(1,  figsize=(16,4))
for i in range(10):  # Plotting paths for 10 simulations
    plt.plot(portfolio_values[:, i], lw=1)
plt.title('Monte Carlo Simulation of Portfolio Prices')
plt.xlabel('Days')
plt.ylabel('Portfolio Value')
plt.grid(True)
plt.show()
```

```python
# Step 7: Distribution of final portfolio values
plt.figure(figsize=(12.5, 6))
plt.hist(final_portfolio_values, bins=100, color='blue', alpha=0.7, edgecolor='black')
plt.axvline(VaR_95, color='red', linestyle='--', label=f'VaR (5%): {VaR_95:.2f}')
plt.title('Distribution of Final Portfolio Values')
plt.xlabel('Portfolio Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 5: Calculate Value at Risk (VaR) as a Percentage
initial_portfolio_value = np.sum(data.iloc[-1] * weights)  # Current portfolio value based on weights
VaR_95_percentage = (VaR_95 / initial_portfolio_value)   # VaR as a percentage of the initial portfolio value
print(f"Value at Risk (5%): {VaR_95_percentage:.2f}")
```

```python
# Example: Assume `data` contains daily returns of 11 stocks in the portfolio
# Replace this with your actual ETF portfolio data
# data = pd.read_csv('your_etf_data.csv')

# For demonstration: Generating random daily returns for 11 stocks
np.random.seed(42)
data = pd.DataFrame(
    np.random.normal(0.001, 0.02, (252, 11)),  # 252 trading days, 11 stocks
    columns=[f"Stock_{i+1}" for i in range(11)]
)

# Define equal weights for all stocks in the portfolio
weights = np.full(len(data.columns), 1 / len(data.columns))

# Step 1: Calculate Portfolio Mean and Volatility
mean_returns = data.mean()  # Mean daily returns for each stock
cov_matrix = data.cov()     # Covariance matrix of stock returns

# Portfolio Mean Return
portfolio_mean = np.dot(weights, mean_returns)

# Portfolio Volatility
portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))

# Annualize the metrics
portfolio_mean_annualized = portfolio_mean * 252
portfolio_volatility_annualized = portfolio_volatility * np.sqrt(252)

# Print the portfolio performance metrics
print("Portfolio Performance Metrics:")
print(f"Mean Daily Return: {portfolio_mean:.5f}")
print(f"Annualized Mean Return: {portfolio_mean_annualized:.2%}")
print(f"Daily Volatility (Std Dev): {portfolio_volatility:.5f}")
print(f"Annualized Volatility: {portfolio_volatility_annualized:.2%}")

# Step 2: Stress Testing
stress_scenarios = {
    "Market Crash": -0.15,       # 15% drop across all stocks
    "Energy Sector Shock": -0.25, # 25% drop for energy-related stocks (assume first 5 stocks)
    "Volatility Spike": 2.0      # Double the volatility of the portfolio
}

# Stress testing results storage
stress_results = {}

for scenario, impact in stress_scenarios.items():
    if scenario == "Market Crash":
        # Uniform negative shock across all stocks
        stressed_data = data + impact
        stressed_returns = stressed_data.dot(weights)
    elif scenario == "Energy Sector Shock":
        # Assume first 5 stocks are energy-related and apply shock only to them
        shocked_data = data.copy()
        shocked_data.iloc[:, :5] *= (1 + impact)
        stressed_returns = shocked_data.dot(weights)
    elif scenario == "Volatility Spike":
        # Increase volatility by scaling standard deviation
        stressed_returns = np.random.normal(
            portfolio_mean, portfolio_volatility * impact, len(data)
        )
    
    # Store results for the scenario
    stress_results[scenario] = stressed_returns

# Step 3: Visualize Stress Testing Results
plt.figure(figsize=(10, 6))

for scenario, returns in stress_results.items():
    plt.hist(returns, bins=50, alpha=0.6, label=scenario)

plt.title("Stress Testing: Portfolio Returns Under Different Scenarios")
plt.xlabel("Portfolio Return")
plt.ylabel("Frequency")
plt.legend()
plt.grid()
plt.show()

# Step 4: Key Metrics Under Stress Scenarios
for scenario, returns in stress_results.items():
    print(f"Scenario: {scenario}")
    print(f"  Mean Return: {np.mean(returns):.2%}")
    print(f"  Standard Deviation: {np.std(returns):.2%}")
    print(f"  Worst Case Return (5th Percentile): {np.percentile(returns, 5):.2%}\n")
```
